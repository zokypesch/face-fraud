```python
import urllib.request
import csv
```

```python
path_source = "../fraud_data_set/prakerja/LCC_FASD_evaluation/spoof"
file_csv_source = "./csv/fake_800_1.csv"
```

```python
def download_images(url):
    file_name = id_generator(12)
    urllib.request.urlretrieve(url, f'{path_source}/{file_name}.jpeg')
    print(f"download successfully {path_source}/{file_name}.jpeg")
    return f'{file_name}.jpeg'
```

```python
import string
import random
import threading
def id_generator(size=6, chars=string.ascii_uppercase + string.digits):
    return ''.join(random.choice(chars) for _ in range(size))
```

```python
list_files = []
with open(file_csv_source) as csv_file:
    # Create a CSV Reader object
    reader = csv.reader(csv_file)
    # Set a line count so we know which line has the headings
    line_count = 0
    for row in reader:
        # Print out the headings from the first line of the file
        if line_count == 0:
            print(f'The column headings are {", ".join(row)}')
            line_count += 1
        # Print out the data from the remaining lines
        else:
            list_files.append(row[2])
            print(f"append successfully: {row[2]}")
            line_count += 1
            
    print(f'CSV File contains {line_count} lines.')
```

```python
import concurrent.futures
values_list = []
with concurrent.futures.ThreadPoolExecutor() as executor:
    futures = []
    for url in list_files:
        futures.append(executor.submit(download_images, url=url))
    for future in concurrent.futures.as_completed(futures):
        values_list.append(future.result())
```
